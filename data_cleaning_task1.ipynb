```python
import pandas as pd
df = pd.read_excel(r"C:\Users\anagh\Downloads\Housing.xlsx.xls")
df
```

```python
df= df.drop_duplicates()
df
```

```python
df = df.copy()
df['price'] = df['price'].astype(str).str.replace(r'[^\d]', '', regex=True).astype(int)
```

```python
df['status'] = df['status'].str.replace(r'[_\.â€¦]+', '', regex=True)
df
```

```python
columns_to_fix = ['guestroom', 'airconditioning', 'basement']
for col in columns_to_fix:
    df[col] = df[col].replace({'y': 'yes', 'n': 'no'})
for col in columns_to_fix:
    print(f"{col} unique values:", df[col].unique())
df
```

```python
df.to_csv("cleaned_housing_data.csv", index=False)
```
